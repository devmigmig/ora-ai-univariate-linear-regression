```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                       
import ipywidgets as widgets

from IPython.display import display, clear_output
```

```python
# Generate synthetic data
np.random.seed(0)
X = np.linspace(0,10,100)
actual_Y = 2 * X + 1 + np.random.normal(0,2,100)
```

```python
# Create sliders
slope_slider = widgets.FloatSlider(
    value=1.8,
    min=-5.0,
    max=5.0,
    step=0.1,
    description='Slope'
)

intercept_slider = widgets.FloatSlider(
    value=2.4,
    min=-10.0,
    max=10.0,
    step=0.1,
    description='Intercept'
)

# Input widgets
x_input = widgets.FloatText(value=5.0, description='X')
y_input = widgets.FloatText(description='Y')

output = widgets.Output()

# Track last edited field
last_changed = {'field': 'x'}
```

```python
# Detect X changes
def on_x_change(change):
    last_changed['field'] = 'x'
    update()

# Detect Y changes
def on_y_change(change):
    last_changed['field'] = 'y'
    update()

# update function
def update(change=None):

    with output:
        clear_output(wait=True)

        slope = slope_slider.value
        intercept = intercept_slider.value

         # Prevent recursive updates
        x_input.unobserve(on_x_change, names='value')
        y_input.unobserve(on_y_change, names='value')

        # Update corresponding value
        if last_changed['field'] == 'x':
            y_input.value = slope * x_input.value + intercept

        elif last_changed['field'] == 'y' and slope != 0:
            x_input.value = (y_input.value - intercept) / slope

        # Re-enable observers
        x_input.observe(on_x_change, names='value')
        y_input.observe(on_y_change, names='value')

        # Regression line
        predicted_Y = slope * X + intercept
        loss = np.mean((predicted_Y - actual_Y) ** 2)

        plt.figure(figsize=(20,10))
        plt.scatter(X, actual_Y, label='Data Points')
        plt.scatter(X, predicted_Y, color='red', label='Linear Regression Line / Predicted Y')

        # Predicted Point (your slope/intercept line)
        plt.scatter(
            x_input.value,
            y_input.value,
            color='green',
            label=f'Predicted Y ({x_input.value:.2f}, {y_input.value:.2f})'
        )

        plt.xlabel('X [Feature]')
        plt.ylabel('Actual Y [Blue]   /   Predicted Y [Red] ')
        plt.title(f'Linear Regression (Loss: {loss:.2f})')
        plt.legend()
        x_max = max(10, x_input.value + 1)
        plt.xticks(np.arange(0, x_max + 1, 1))
        plt.yticks(np.arange(0, 45, 2))   
        plt.grid(True)
        plt.show()
```

```python
# Observe widget changes
x_input.observe(on_x_change, names='value')
y_input.observe(on_y_change, names='value')
slope_slider.observe(update, names='value')
intercept_slider.observe(update, names='value')

# Display everything
display(
    slope_slider,
    intercept_slider,
    x_input,
    y_input,
    output
)

# Initial graph
update()
```
